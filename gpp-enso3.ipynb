```python
import xarray as xr
import matplotlib.pyplot as plt

# 读取文件
ds = xr.open_dataset("D:\cangku\data\tos_Omon_CanESM5_G1_r1i1p2f1_gn_185001-194912.nc")

# 看看坐标信息
print(ds)

# 获取二维经纬度
lat = ds["latitude"]
lon = ds["longitude"]

print("lat shape:", lat.shape)
print("lon shape:", lon.shape)

# 取一个时间片的 tos 数据
tos = ds["tos"].isel(time=0)  # 第一个时间点

# 作图
plt.figure(figsize=(10, 5))
plt.pcolormesh(lon, lat, tos, shading="auto")  # 使用二维 lat/lon
plt.colorbar(label="tos (K)")  # 温度变量，单位通常是 K
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("CanESM5 tos (time=0)")
plt.show()
```
